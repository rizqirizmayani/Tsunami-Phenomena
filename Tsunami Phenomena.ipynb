## Background

Bencana alam adalah kejadian yang sudah sering terjadi. Bencana alam terjadi karena serangan iklim dan meteorologi yang terjadi sangat kuat dan tiba-tiba yang menghasilkan kejadian-kejadian seperti badai salju, banjir, angin topan, tornado dan lain-lain. Selain itu tanah longsor, gempa bumi, tsunami dan letusan gunung api juga termasuk dalam bencana alam. Kejadian-kejadian tersebut tidak terjadi disembarang tempat, tetapi cenderung hanya di daerah-daerah tertentu saja. Kejadian tersebut bisanya terjadi dimulai dengan sangat cepat dan tidak terduga, walaupun dengan kemajuan teknologi dan ilmu pengetahuan saat ini sudah bisa memprediksi bencana alam yang akan terjadi. Namun, gempa, banjir, tsunami dan tornado datang secara tiba-tiba (Baum dkk, 2002). Tsunami adalah gelombang, atau serangkaian gelombang dalam “kereta ombak”, yang dihasilkan oleh perpindahan vertikal kolom air secara tiba-tiba. Perpindahan ini dapat disebabkan oleh aktivitas seismik, vulkanisme eksplosif, tanah longsor di atas atau di bawah air, dampak asteroid atau fenomena meteorologi tertentu. Gelombang ini dapat dihasilkan di lautan, teluk, danau, sungai atau reservior. Istilah tsunami berasal dari bahasa Jepang yang mempunyai arti pelabuhan (tsu) gelombang (nami), karena gelombang seperti itu sering berkembang sebagai fenomena resonansi di pelabuhan setelah gempa bumi di lepas pantai (Bryant, 2014).

```python
df['EQ_MAGNITUDE'].fillna(df['EQ_MAGNITUDE'].median(), inplace=True)
```

Terdapat beberapa missing value pada kolom magnitude dan penulis ingin mengisi missing value tersebut. Langkah awal yang dilakukan adalah melihat persebaran data dengan menggunakan histogram dan perhitungan skew. Data magnitude menunjukan persebaran data yang tidak normal, sehingga untuk mengisi missing value menggunakan nilai median.
Pemberian nilai pada missing value berdasarkan konsep terbentuknya tsunami, yang mana tsunami terjadi ketika magnitude mencapai nilai tertentu.

```python
df['kategory_tahun']=0
for i in df.index:
  if df["YEAR"][i]<1980:
    df["kategory_tahun"][i]='70s 20th century'
  elif (df["YEAR"][i]>=1980) & (df["YEAR"][i] < 1990):
    df["kategory_tahun"][i]='80s 20th century'
  elif (df["YEAR"][i]>=1990) & (df["YEAR"][i] < 2000):
    df["kategory_tahun"][i]='90s 20th century'
  elif (df["YEAR"][i]>=2000) & (df["YEAR"][i] < 2010):
    df["kategory_tahun"][i]='00s 21th century' 
  else:
    df["kategory_tahun"][i]='10s 21th century'
```

Penambahan kolom baru data kategori kelompok tahun dengan menggunakan rentang 10 tahun. Pembagian ini selanjutnya akan digunakan untuk visualisasi dan uji hipotesis

```python
df['hipotesis']=1
for i in df.index:
  if df["YEAR"][i]<2010:
    df["hipotesis"][i]='h0' 
  else:
    df["hipotesis"][i]='h1'
```

Penambahan kolom baru untuk membagi data menjadi dua sampel, yang selanjutnya akan digunakan untuk uji hipotesis

### Perbedaan nilai magnitudo yang terjadi pada tahun 1970-2010 dengan 10 tahun terakhir

```python
kolom_list2 = ['ID', 'YEAR', 'MONTH', 'DAY', 'LATITUDE', 'LONGITUDE', 'COUNTRY', 'CAUSE', 'DAY', 'EQ_DEPTH', 'kategory_tahun', 'intensitas']
df_hipotesis = df1.drop(kolom_list2, axis=1)
```

```python
df_hipotesis
```

```python
h1 = df_hipotesis[(df_hipotesis['hipotesis']=='h1')]
h0 = df_hipotesis[(df_hipotesis['hipotesis']=='h0')]
```

```python
print ('Mean: ' + str(h0['EQ_MAGNITUDE'].mean()))
print ('Mean: ' + str(h1['EQ_MAGNITUDE'].mean()))
```

```python
# t-test 2 sample 2 tailed
t_stat, p_val = stats.ttest_ind(h0['EQ_MAGNITUDE'], h1['EQ_MAGNITUDE'])

# ttest_ind is 2-tailed, we dont need to divide by 2
print('P-value:',p_val)
print('t-statistics:',t_stat)
```

```python
# simulation
h0_pop = np.random.normal(h0['EQ_MAGNITUDE'].mean(),h0['EQ_MAGNITUDE'].std(),10000)
h1_pop = np.random.normal(h1['EQ_MAGNITUDE'].mean(),h1['EQ_MAGNITUDE'].std(),10000)

# confidence interval based on BALCONS RECYLING
ci = stats.norm.interval(0.95, h0['EQ_MAGNITUDE'].mean(), h0['EQ_MAGNITUDE'].std())

plt.figure(figsize=(16,5))
sns.histplot(h0_pop, label='Magnitude tahun 1970-2010 (POP)',color='blue')
sns.histplot(h1_pop, label='Magnitude tahun 2010-2020 (POP)',color='red')

# mean
plt.axvline(h0['EQ_MAGNITUDE'].mean(), color='yellow', linewidth=2, label='Rerata Magnitude tahun 1970-2010')
plt.axvline(h1['EQ_MAGNITUDE'].mean(), color='magenta',  linewidth=2, label='Rerata Magnitude tahun 2010-2020')

plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%')

# alternative hypotesis
plt.axvline(h0_pop.mean()+t_stat*h0_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Hipotesis Alternatif')
plt.axvline(h1_pop.mean()-t_stat*h0_pop.std(), color='black', linestyle='dashed', linewidth=2)

plt.legend()
print(ci[0])
print(ci[1])
```

- H0 = Rerata magnitudo 10 tahun terakhir sama dengan rerata magnitudo tahun 1970-2010
- H1 = Rerata magnitudo 10 tahun terakhir tidak sama dengan rerata magnitudo tahun 1970-2010

Hasil t_test menunjukan nilai p-value sebesar 0.62726. Hasil tersebut menunjukan H0 gagal ditolak, dengan arti tidak terdapatnya perbedaan yang signifikan diantara dua sampel. Grafik juga menunjukan garis hipotesis terbentuk diantara CI dan deka dengan nilai mean antara kedua populasi yang memiliki arti H0 gagal ditolak.
